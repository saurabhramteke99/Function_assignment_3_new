Que_5  What are iterators in Python and how do they differ from iterables?

Ans:  1. Iterable:

An iterable is any Python object capable of returning its members one at a time, allowing it to be iterated over (looped through). In simple terms, an iterable is something you can use with a for loop or pass to functions like iter() to create an iterator.

An iterable object implements the __iter__() method, which returns an iterator. Common examples of iterables in Python include lists, tuples, dictionaries, and sets.

2. Iterator:
An iterator is an object that represents a stream of data; it keeps track of the current position while iterating through a collection. It is the object that performs the actual iteration, returning one element at a time when next() is called on it. An iterator is created from an iterable by calling the iter() function.

An iterator implements two methods:

__iter__(): Returns the iterator object itself. This allows an iterator to be used in a for loop or other places that expect an iterable.

__next__(): Returns the next item from the collection. If there are no more items, it raises a StopIteration exception to signal that the iteration is complete.


Que_6. Explain the concept of generators in Python and how they are defined.

Ans: Definition: Generators are a simple way to create iterators using functions and the yield keyword instead of returning values.

Yield: Pauses the function and sends a value to the caller. The state of the function (including local variables) is saved. The function can then resume execution where it left off

Key Characteristics of Generators:

a) Lazy Evaluation:
Generators produce values one at a time as you iterate over them. They do not compute or store all values in memory at once.

b) State Preservation:
Generators maintain their internal state between iterations, which allows them to remember where they left off.

c) Efficiency:
Since they generate values only when needed (i.e., on-demand), they are more memory-efficient than other data structures that store all values at once, like lists.

d) Automatically Closed:
Once a generator function finishes executing or raises a StopIteration exception, it is closed automatically.



Que_7 What are the advantages of using generators over regular functions?

Ans: Using generators over regular functions in Python offers several key advantages. These advantages are mainly due to the lazy evaluation and memory efficiency that generators provide. Below are the main benefits of using generators:

1. Memory Efficiency:

a)Generators are more memory-efficient compared to regular functions that return full sequences (like lists, tuples, etc.). This is because generators only produce one value at a time and do not store the entire sequence in memory.

 For example, when dealing with large datasets or infinite sequences, a regular function would require a lot of memory to hold all values at once, while a generator only holds one value in memory at any given time.

2.Lazy Evaluation (On-Demand Computation):

a)Lazy evaluation means that values are only computed when requested, rather than upfront. This is especially useful when you are working with potentially infinite sequences or large datasets where you only need one item at a time.

b)A regular function that returns a list or other collection will calculate and store all values at once, whereas a generator produces each value only when it is needed, saving both time and resources.

Que_8 What is a lambda function in Python and when is it typically used?

Ans: Definition: Lambda functions are small, anonymous functions defined using the lambda keyword. They are used for creating small, throwaway functions without the need to formally define a function using def.

When to Use Lambda Functions:

Lambda functions are useful in scenarios where you need a short function for a specific task, especially when the function is not going to be reused elsewhere in your code. They are commonly used in the following cases:

1)Short, Throwaway Functions: When you need a function for a short-term, specific purpose (for example, in a map, filter, or reduce operation), lambda functions are a good choice. Instead of defining a full function using def, a lambda function allows you to define a simple operation in one line.

2)Functions Used as Arguments: Lambda functions are often passed as arguments to functions that require a function as a parameter, like map(), filter(), sorted(), and reduce().

3)Inline Functions: If you need to define a function in a small scope and don't want to bother with defining a full function using def, lambda functions are a convenient alternative.

4)Simplifying Code: Lambda functions can make your code more concise, especially in places where a function is needed only once and does not need to be reused elsewhere.

Que_10  What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?

Ans: In Python, the functions map(), reduce(), and filter() are all used to apply a function to elements of an iterable, but they differ in their behavior and purpose. Here's a breakdown of each function:

1) map():


Purpose: The map() function is used to apply a given function to each item in an iterable (such as a list or tuple) and return a map object (an iterator) containing the results.

Behavior: It processes each element of the iterable individually and returns a transformed version of the iterable.

Usage: It's typically used for transforming or modifying each item in an iterable.

2)reduce()

Purpose: The reduce() function from the functools module is used to apply a function cumulatively to the items of an iterable, reducing the iterable to a single accumulated result.

Behavior: It applies the function to the first two elements of the iterable, then applies the function to the result and the next element, and so on, until only one value is left.

Usage: It's typically used when you need to combine all elements of an iterable into a single value (like summing, multiplying, finding the maximum, etc.).


3)filter()

Purpose: The filter() function is used to filter the items of an iterable by applying a function that returns either True or False. Only the elements that return True will be included in the result.

Behavior: It processes each element in the iterable and retains only those for which the function returns True.

Usage: It's typically used for filtering elements from an iterable based on some condition.

````python
# Que_4. Write a Python function that checks if a given number is prime or not from 1 to 200.

def is_prime_num(n):
  if n <= 1:
    return False

  for i in range(2,int(n**0.5)+ 1):

    if i % n == 0:
      return False


  return True
````

````python
# Que_5.  Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
# terms.

class FibonacciIterator:
    def __init__(self, terms):
        self.terms = terms  # Number of terms to generate
        self.a, self.b = 0, 1  # First two Fibonacci numbers
        self.count = 0  # Counter to track the number of terms generated

    def __iter__(self):
        return self  # The iterator object itself

    def __next__(self):
        if self.count < self.terms:
            fib_number = self.a
            self.a, self.b = self.b, self.a + self.b  # Update to the next Fibonacci numbers
            self.count += 1
            return fib_number
        else:
            raise StopIteration  # Stop iteration once the required terms are generated

# Example usage:
fibonacci = FibonacciIterator(10)

for num in fibonacci:
    print(num)
````

````python
#Que_7. Implement a generator function that reads a file line by line and yields each line as a string.

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()  # Strip removes any leading/trailing whitespace, including newlines
````

````python
# Assuming the file is in the same directory as the script


file_path = Hello, World!/n python is great/n Generators are awosome!

# Using the generator to print each line from the file
for line in read_file_line_by_line(file_path):
    print(line)


````
